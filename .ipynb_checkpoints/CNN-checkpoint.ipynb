```
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout

batch_size = 128
n_classes = 10
epochs = 10

(x_train, y_train), (x_test, y_test) = mnist.load_data()
print(x_train.shape)
# 28*28の画像が60000枚

# CNNはチャンネル数が明示されていないと動かない
# reshape方法その1
# x_train = x_train.reshape(60000, 28, 28, 1)
# reshape方法その２
x_train = x_train.reshape(-1, 28, 28, 1)
print(x_train.shape)
x_test = x_test.reshape(-1, 28, 28, 1)
# print(x_train[0])

# float型に変換しておく
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')

# 最大値で割って割合にする
x_train /= 255
x_test /= 255

# 続いてyの前処理
print(y_train)
# to_categoricalで0か1かに変換してくれる
y_train = keras.utils.to_categorical(y_train, n_classes)
y_test = keras.utils.to_categorical(y_test, n_classes)
print(y_train)
```

```
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(10, 15))

for i in range(100):
  plt.subplot(10, 10, i + 1, xticks = [], yticks = [])
  plt.title(y_train[i].argmax())
  plt.axis('off')
  plt.imshow(x_train[i].reshape((28, 28)), cmap = 'gray')
```

```
from keras.layers.convolutional import Conv2D
from keras.layers.pooling import MaxPool2D
from keras.layers.core import Activation, Flatten

model = Sequential()
# 10はチャンネル数
model.add(Conv2D(10, (3, 3), padding = 'same', input_shape = (28, 28, 1)))
model.add(Activation('relu'))
model.add(Conv2D(16, (3, 3), padding = 'same'))
model.add(Activation('relu'))
model.add(MaxPool2D())
# ConvとPoolを繰り返す

model.add(Flatten())
model.add(Dense(n_classes, activation = 'softmax'))
model.compile(optimizer = 'sgd', loss = 'categorical_crossentropy', metrics = ['accuracy'])

model.summary()
```

```
model.fit(x_train, y_train, batch_size = batch_size, epochs = epochs,
          verbose = 1,#途中経過を出力する（0: 出力しない,1: 出力する）,
          validation_split = 0.2)# 20%のデータは検証用に使う

score = model.evaluate(x_test, y_test)
print('print loss = ', score[0])
print('print accuracy = ', score[1])
```
