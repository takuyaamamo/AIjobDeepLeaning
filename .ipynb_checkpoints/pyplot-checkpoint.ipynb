```
import numpy as np
import matplotlib.pyplot as plt

n_cols = 2
n_rows = 2

# figureで図を作成figsizeは縦8インチ横20インチの意味
plt.figure(figsize= (20, 8))

x = np.arange(-5, 5 , 0.1)
y = x

# subplotで図を4箇所に分け2行2列で表示するこれは1個目
plt.subplot(n_rows, n_cols, 1)
plt.title('x')
plt.ylim(-10, 10)
plt.plot(x, y)

y = -2 * x - 3 * x + x
plt.subplot(n_rows, n_cols, 2)
plt.title('2x - 3x + x')
plt.ylim(-10, 10)
plt.plot(x, y)

y = np.sin(x)
plt.subplot(n_rows, n_cols, 3)
plt.title('sin(x)')
plt.ylim(-10, 10)
plt.plot(x, y)

y = np.exp(x)
plt.subplot(n_rows, n_cols, 4)
plt.title('exp(x)')
plt.ylim(-10, 10)
plt.plot(x, y)
```
