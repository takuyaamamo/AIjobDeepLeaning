```
# クラスの宣言
class LSTM(object):
#   コンストラクタ、クラスの最初には”__init__”でクラスの中で使用する変数を定義する必要がある
    def __init__(self, input_size, hidden_size, output_size):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size

      # init weight（入力層）の定義
      # プレースホルダーはデータが格納される入れ物。データは未定のままグラフを構築し、具体的な値は実行する時に与える。
        self.inputs = tf.placeholder(tf.float32, shape=[None, None, self.input_size], name='inputs')

        
      # 忘却ゲートの定義
      # tf.zerosでinput_size行hidden_size列にゼロの行列を作成し初期化
      # Variableで変数の定義
        self.Wf = tf.Variable(tf.zeros([input_size, hidden_size])) # 新規入力の個数
        self.Uf = tf.Variable(tf.zeros([hidden_size, hidden_size])) # 前回の出力を入力の個数
        self.bf = tf.Variable(tf.zeros([hidden_size])) #バイアスの個数
 
      # 入力ゲート（忘却ゲートのコピペ）
        self.Wi = tf.Variable(tf.zeros([input_size, hidden_size])) # 新規入力の個数
        self.Ui = tf.Variable(tf.zeros([hidden_size, hidden_size])) # 前回の出力を入力の個数
        self.bi = tf.Variable(tf.zeros([hidden_size])) #バイアスの個数
      
      # セルへの書き込み部分（忘却ゲートのコピペ）
        self.Wc = tf.Variable(tf.zeros([input_size, hidden_size])) # 新規入力の個数
        self.Uc = tf.Variable(tf.zeros([hidden_size, hidden_size])) # 前回の出力を入力の個数
        self.bc = tf.Variable(tf.zeros([hidden_size])) #バイアスの個数
 
        
      # 出力ゲート（忘却ゲートのコピペ）
        self.Wo = tf.Variable(tf.zeros([input_size, hidden_size])) # 新規入力の個数
        self.Uo = tf.Variable(tf.zeros([hidden_size, hidden_size])) # 前回の出力を入力の個数
        self.bo = tf.Variable(tf.zeros([hidden_size])) #バイアスの個数

        
      # 出力層（忘却ゲートのコピペ）
      # tf.truncated_normalはTensorを正規分布かつ標準偏差の２倍までのランダムな値で初期化する
        self.Wv = tf.Variable(tf.truncated_normal([hidden_size, output_size], mean=0, stddev=0.01)) # 最後の出力の個数。出力したいサイズに合わせる
        self.bv = tf.Variable(tf.truncated_normal([output_size], mean=0, stddev=0.01))# 最後の出力の個数。出力したいサイズに合わせる
        
      # 初期に使用する前回の出力を定義tf.matmulは行列積を行う
        self.init_hidden = tf.matmul(self.inputs[:,0,:], tf.zeros([input_size, hidden_size]))
      # tf.stackくっつけて並べ替えている
        self.init_hidden = tf.stack([self.init_hidden, self.init_hidden])

        self.input_fn = self._get_batch_input(self.inputs)

    def forward(self, prev_state, x):
#    前回の出力から前回の計算を取り出してくれる
        h, c = tf.unstack(prev_state)

#   忘却ゲートの計算
        f = tf.sigmoid(tf.matmul(x, self.Wf) + tf.matmul(h, self.Uf) + self.bf)
#   入力ゲートの計算
        i = tf.sigmoid(tf.matmul(x, self.Wi) + tf.matmul(h, self.Ui) + self.bi)
#  セルへの書き込み部分の計算
        c_tmp = tf.nn.tanh(tf.matmul(x, self.Wc) + tf.matmul(h, self.Uc) + self.bc)
#  出力ゲートの計算
        o = tf.sigmoid(tf.matmul(x, self.Wo) + tf.matmul(h, self.Uo) + self.bo)
    
 #   忘却ゲートのCへの掛け算
        c = f * c #忘却ゲートによってメモリセルCから情報を消す
#    入力ゲートのCへの足し算
        c = c + i * c_tmp #情報を書き込む
#    出力ゲートの計算
        h = o * tf.nn.tanh(c)

#    一旦合体、使う時に分解する
        return tf.stack([h, c])


    def _get_batch_input(self, inputs):
#       tf.transpose転置テンソルを行うperm を指定すると指定されたパラメータにしたがって添字を入れ替えてくれます。
        return tf.transpose(tf.transpose(inputs, perm=[2, 0, 1]))

    def calc_all_layers(self):
        all_hidden_states = tf.scan(self.forward, self.input_fn, initializer=self.init_hidden, name='states')
        return all_hidden_states[:, 0, :, :]

    def calc_output(self, state):
        return tf.nn.relu(tf.matmul(state, self.Wv) + self.bv)

    def calc_outputs(self):
        all_states = self.calc_all_layers()
        all_outputs = tf.map_fn(self.calc_output, all_states)

        return all_outputs
```

```
def get_on_hot(number):
    on_hot=[0]*10
    on_hot[number]=1
    return on_hot

#  .load_digits()は MNISTよりも少し小さな手書き文字認識のためのデータセット
digits = datasets.load_digits()
X=digits.images
Y_=digits.target

Y=map(get_on_hot,Y_)

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

X_train=X_train[:1400]
y_train=y_train[:1400]
```

```
from IPython.display import clear_output
  
hidden_size = 50
input_size = 8
output_size = 10

y = tf.placeholder(tf.float32, shape=[None, output_size], name='inputs')

lstm = LSTM(input_size, hidden_size, output_size)
outputs = lstm.calc_outputs()

last_output = outputs[-1]

output = tf.nn.softmax(last_output)

# cross entropy loss
loss = -tf.reduce_sum(y * tf.log(output))

train_step = tf.train.AdamOptimizer().minimize(loss)

correct_predictions = tf.equal(tf.argmax(y, 1), tf.argmax(output, 1))
acc = (tf.reduce_mean(tf.cast(correct_predictions, tf.float32)))

sess=tf.InteractiveSession()
sess.run(tf.global_variables_initializer())

log_loss = []
log_acc = []
log_val_acc = []

for epoch in range(100):
    
    start=0
    end=100
    for i in range(14):
        
        X=X_train[start:end]
        Y=y_train[start:end]
        
        start=end
        end=start+100
        sess.run(train_step,feed_dict={lstm.inputs:X, y:Y})
    
    log_loss.append(sess.run(loss,feed_dict={lstm.inputs:X, y:Y}))
    log_acc.append(sess.run(acc,feed_dict={lstm.inputs:X_train[:500], y:y_train[:500]}))
    log_val_acc.append(sess.run(acc,feed_dict={lstm.inputs:X_test, y:y_test}))
    
    clear_output(wait=True)
    
    plt.subplot(1, 2, 1)
    plt.plot(range(epoch+1), log_loss)
    
    plt.subplot(1, 2, 2)
    plt.ylim(0., 1.)
    plt.plot(range(epoch+1), log_acc)
    plt.ylim(0., 1.)
    plt.plot(range(epoch+1), log_val_acc)
    
    plt.show()
    
    print("\r[%s] loss: %s acc: %s val acc: %s"%(epoch, log_loss[-1], log_acc[-1], log_val_acc[-1])),
```
